```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
pd.options.plotting.backend = 'plotly'
```

```python
df_pc = pd.read_csv('../../Data/1hour_concat.csv', sep=',')
df_pc.head()
```

```python
#Sort by Date, then by Canton
df_pc = df_pc.sort_values(by=['Date', 'Canton'])
df_pc = df_pc.reset_index(drop=True)
df_pc['Consumption'] = df_pc['Consumption'] * -1
df_pc.head()
```

```python
df_pivot = df_pc.pivot(index='Date', columns='Canton', values='Production').reset_index()
df_pivot.columns = [str(col) + '_prod' for col in df_pivot.columns]

df_pivot.head()
```

```python
dropdown_options = []
for canton in df_pivot.columns:
    dropdown_options.append({'label': canton, 'value': canton})

print(dropdown_options)
```
