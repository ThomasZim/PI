```python
import pandas as pd
import matplotlib as plty
import plotly.express as px
pd.options.plotting.backend = 'matplotlib'
```

```python
printGraphs = False
```

```python
#Open dataset
df = pd.read_csv('../../Data/ogd104_stromproduktion_swissgrid.csv', sep=',')

#Create a new dataframe for every different type of energy
df_wind = df[df['Energietraeger'] == 'Wind']
df_thermal = df[df['Energietraeger'] == 'Thermische']
df_storage = df[df['Energietraeger'] == 'Speicherkraft']
df_photovoltaic = df[df['Energietraeger'] == 'Photovoltaik']
df_nuclear = df[df['Energietraeger'] == 'Kernkraft']
df_flow = df[df['Energietraeger'] == 'Flusskraft']

#Reindex the dataframes
df_wind = df_wind.reset_index(drop=True)
df_wind.head()

#Make sure every dataframe length added up is equal to the original dataframe
print(len(df_wind) + len(df_thermal) + len(df_storage) + len(df_photovoltaic) + len(df_nuclear) + len(df_flow))
print(len(df))
```

# Normal plot using pandas

```python
#Plot all the dataframes in one plot using pandas

if printGraphs:
  df_wind.plot(kind='line', x='Datum', y='Produktion_GWh', 
            color='blue', grid=True, 
            title='Wind energy production in Switzerland')

  df_thermal.plot(kind='line', x='Datum', y='Produktion_GWh',
                color='red', grid=True,
                  title='Thermal energy production in Switzerland')

  df_storage.plot(kind='line', x='Datum', y='Produktion_GWh',
                  color='green', grid=True,
                  title='Storage energy production in Switzerland')

  df_photovoltaic.plot(kind='line', x='Datum', y='Produktion_GWh',
                          color='yellow', grid=True,
                          title='Photovoltaic energy production in Switzerland')

  df_nuclear.plot(kind='line', x='Datum', y='Produktion_GWh',
                  color='orange', grid=True,
                  title='Nuclear energy production in Switzerland')

  df_flow.plot(kind='line', x='Datum', y='Produktion_GWh',
                  color='purple', grid=True,
                  title='Flow energy production in Switzerland')
```
